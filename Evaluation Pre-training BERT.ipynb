```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
history_df_01 = pd.read_csv('custom-bert-history-01.csv')
history_df_02 = pd.read_csv('custom-bert-fine-tuned-history-01.csv')
```

```python
history_pretrain_df = pd.concat([history_df_01, history_df_02])
history_pretrain_df = history_pretrain_df.reset_index(drop=True)
```

```python
history_pretrain_df = history_pretrain_df.rename(columns={'loss': 'Training Loss after Custom Pre-training BERT', 
                                        'val_loss': 'Validation Loss after Custom Pre-training BERT'})
```

```python
history_pretrain_df
```

```python
history_pretrain_df = history_pretrain_df[['Training Loss after Custom Pre-training BERT', 
                         'Validation Loss after Custom Pre-training BERT']]
```

```python
history_pretrain_df.head()
```

```python
plt.figure(figsize=(10, 8))
ax = sns.lineplot(data=history_pretrain_df)
ax.set(xlabel='epochs', ylabel='loss')
plt.show()
```

```python
history_df_01 = pd.read_csv('../Scenario 1/fine-tuning-bert-history-01.csv')
history_df_02 = pd.read_csv('../Scenario 1/fine-tuning-bert-history-01-fine-tuned.csv')
```

```python
history_finetune_df = pd.concat([history_df_01, history_df_02])
history_finetune_df = history_finetune_df.reset_index(drop=True)
history_finetune_df = history_finetune_df.rename(columns={'loss': 'Training Loss by Fine-tuning normal BERT', 
                                        'val_loss': 'Validation Loss by Fine-tuning normal BERT'})
history_finetune_df
```

```python
history_finetune_df = history_finetune_df[['Training Loss by Fine-tuning normal BERT', 
                         'Validation Loss by Fine-tuning normal BERT']]
```

```python
history_df = pd.concat([history_pretrain_df, history_finetune_df], axis=1)
```

```python
history_df = history_df[['Validation Loss by Fine-tuning normal BERT', 'Validation Loss after Custom Pre-training BERT']]
```

```python
history_df
```

```python
history_df = history_df.rename(columns={'Validation Loss by Fine-tuning normal BERT': 'Fine-tuning standard BERT', 
                                        'Validation Loss after Custom Pre-training BERT': 'Fine-tuning custom pre-trained BERT'})
```

```python
plt.figure(figsize=(10, 8))
plt.rc('axes', labelsize=20)    # fontsize of the x and y labels
plt.rc('legend', fontsize=20)    # legend fontsize
ax = sns.lineplot(data=history_df)
ax.set(xlabel='Epochs', ylabel='Validation Loss')
plt.show()
```
