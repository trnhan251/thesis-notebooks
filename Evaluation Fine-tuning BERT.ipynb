```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# First Training

```python
history_df_01 = pd.read_csv('fine-tuning-bert-history-01.csv')
```

```python
history_df_02 = pd.read_csv('fine-tuning-bert-history-01-fine-tuned.csv')
```

```python
history_df = pd.concat([history_df_01, history_df_02])
history_df = history_df.reset_index(drop=True)
history_df
```

```python
train_df = history_df[['loss']]
```

```python
val_df = history_df[['val_loss']]
```

```python
val_df.head()
```

```python
df = train_df.join(val_df['val_loss'])
```

```python
df.head()
```

```python
df = df.rename(columns={'loss': 'BERT Loss', 'val_loss': 'BERT Validation Loss'})
```

```python
plt.figure(figsize=(10, 8))
ax = sns.lineplot(data=df)
ax.set(xlabel='epochs', ylabel='loss')
ax.set_title('Training model with fine-tuning BERT')
plt.show()
```

# Combine with ALBERT

```python
history_df_01 = pd.read_csv('fine-tuning-albert-history-01.csv')
history_df_02 = pd.read_csv('fine-tuning-albert-history-01-fine-tuned.csv')
```

```python
history_df = pd.concat([history_df_01, history_df_02])
history_df = history_df.reset_index(drop=True)
history_df
```

```python
train_df = history_df[['loss']]
val_df = history_df[['val_loss']]
df_2 = train_df.join(val_df['val_loss'])
df_2.head()
```

```python
df_2 = df_2.rename(columns={'loss': 'ALBERT Loss', 'val_loss': 'ALBERT Validation Loss'})
```

```python
df_new = pd.concat([df, df_2], axis=1)
```

```python
df_new
```

```python
plt.figure(figsize=(10, 8))
ax = sns.lineplot(data=df_new)
ax.set(xlabel='epochs', ylabel='loss')
ax.set_title('Training model with fine-tuning')
plt.show()
```
