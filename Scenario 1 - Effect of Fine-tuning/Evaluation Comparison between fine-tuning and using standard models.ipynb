```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# BERT Validation Fine-tuned DF

```python
history_df_01 = pd.read_csv('fine-tuning-bert-history-01.csv')
```

```python
history_df_02 = pd.read_csv('fine-tuning-bert-history-01-fine-tuned.csv')
```

```python
history_df = pd.concat([history_df_01, history_df_02])
history_df = history_df.reset_index(drop=True)
history_df
```

```python
bert_val_df_fine_tuned = history_df[['val_loss']]
```

```python
bert_val_df_fine_tuned.head()
```

```python
bert_val_df_fine_tuned = bert_val_df_fine_tuned.rename(columns={'val_loss': 'Fine-tuning BERT'})
```

```python
bert_val_df_fine_tuned.head()
```

# ALBERT Validation Fine-tuned DF

```python
history_df_01 = pd.read_csv('fine-tuning-albert-history-01.csv')
history_df_02 = pd.read_csv('fine-tuning-albert-history-01-fine-tuned.csv')
```

```python
history_df = pd.concat([history_df_01, history_df_02])
history_df = history_df.reset_index(drop=True)
```

```python
albert_val_df_fine_tuned = history_df[['val_loss']]
albert_val_df_fine_tuned = albert_val_df_fine_tuned.rename(columns={'val_loss': 'Fine-tuning ALBERT'})
```

```python
albert_val_df_fine_tuned.head()
```

# BERT Validation Standard DF

```python
history_df_01 = pd.read_csv('using-standard-bert-history-01.csv')
history_df_01.head()
```

```python
bert_val_df_standard = history_df_01[['val_loss']]
bert_val_df_standard = bert_val_df_standard.rename(columns={'val_loss': 'Only using BERT as feature extractor'})
```

```python
bert_val_df_standard.head()
```

# ALBERT Validation Standard DF

```python
history_df_01 = pd.read_csv('using-standard-albert-history-01.csv')
history_df_01.head()
```

```python
albert_val_df_standard = history_df_01[['val_loss']]
albert_val_df_standard = albert_val_df_standard.rename(columns={'val_loss': 'Only using ALBERT as feature extractor'})
```

```python
albert_val_df_standard.head()
```

# Combining

```python
df = pd.concat([bert_val_df_standard, bert_val_df_fine_tuned, albert_val_df_standard, albert_val_df_fine_tuned], axis=1)
```

```python
df
```

```python
plt.figure(figsize=(10, 8))

plt.rc('axes', labelsize=14)    # fontsize of the x and y labels
plt.rc('legend', fontsize=14)    # legend fontsize
ax = sns.lineplot(data=df)
ax.set(xlabel='Epochs', ylabel='Validation Loss')
plt.show()
```
