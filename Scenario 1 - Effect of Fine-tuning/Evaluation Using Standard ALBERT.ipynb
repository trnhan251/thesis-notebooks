```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# First Training

```python
history_df_01 = pd.read_csv('using-standard-albert-history-01.csv')
```

```python
history_df_01
```

```python
train_df = history_df_01[['loss']]
```

```python
val_df = history_df_01[['val_loss']]
```

```python
val_df.head()
```

```python
df = train_df.join(val_df['val_loss'])
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 8))
ax = sns.lineplot(data=df)
ax.set(xlabel='epochs', ylabel='mean squared error')
ax.set_title('Training model using standard ALBERT')
plt.show()
```
